```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
plt.grid()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot.

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
ax.set(xlabel = "Date", ylabel = "Fare($)", title ="PyBer Fare by Month")
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)

# Add a grid.
ax.grid()
# Add the legend.
ax.legend()

```

## Scatter plots

```python
 # Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, color='r', label="Chicago")
plt.xlabel('Fare($)')
plt.ylabel('Month')
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.legend(loc='upper right')
```

```python
plt.scatter(y_axis, x_axis, s=y_axis)
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(y_axis, x_axis, s=y_axis_larger, color='skyblue', alpha=0.8, edgecolor='black', linewidths=2, label="Boston")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.xlim(0, 50)
plt.legend()
plt.xlabel('Fare($)')
plt.ylabel('Month')
```

```python


# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
explode = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,labels=x_axis, explode=explode, autopct='%1.1f%%', shadow=True, startangle=80, counterclock=False)
ax.axis('equal')

plt.show()
```
